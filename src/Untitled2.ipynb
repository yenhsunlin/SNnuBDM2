```python
plt.plot(np.linspace(0,np.arccos(5/getMomentum(5,1)),100)/np.pi,
         getEv(5,1,np.linspace(0,np.arccos(5/getMomentum(5,1)),100)),label=r'$T_\chi=5$ MeV')
plt.plot(np.linspace(0,np.arccos(50/getMomentum(50,1)),100)/np.pi,
         getEv(50,1,np.linspace(0,np.arccos(50/getMomentum(50,1)),100)),label=r'$T_\chi=50$ MeV')
plt.plot(np.linspace(0,np.arccos(100/getMomentum(100,1)),30)/np.pi,
         getEv(100,1,np.linspace(0,np.arccos(100/getMomentum(100,1)),30)),label=r'$T_\chi=100$ MeV')
plt.xlim(-0.02,0.25+0.02)
plt.legend()
#plt.ylim(-10,600)
plt.yscale('log')
plt.xlabel(r'Scattering angle $\psi~[\pi]$')
plt.ylabel(r'Required incoming $E_\nu$ [MeV]')
plt.text(0.23,6,r'$m_\chi=1$ MeV',horizontalalignment='center')
plt.savefig('../plots/required_Ev.pdf',bbox_inches='tight')
```

```python
gV=1e-3
gD=gV
Tx1=5
Tx2=50
Tx3=100
mx1 = 1
psi_ls1 = np.linspace(0,np.arccos(Tx1/getMomentum(Tx1,mx1)),100)
psi_ls2 = np.linspace(0,np.arccos(Tx2/getMomentum(Tx2,mx1)),100)
psi_ls3 = np.linspace(0,np.arccos(Tx3/getMomentum(Tx3,mx1)),100)
diffcrox1 = []
diffcrox2 = []
diffcrox3=[]
for psi in psi_ls1:
    diffcrox1.append(diffCrossSectionNuDM(Tx1,mx1,mx,psi,gV,gD))
for psi in psi_ls2:
    diffcrox2.append(diffCrossSectionNuDM(Tx2,mx1,mx,psi,gV,gD))
for psi in psi_ls3:
    diffcrox3.append(diffCrossSectionNuDM(Tx3,mx1,mx,psi,gV,gD))
```

```python
plt.plot(psi_ls1/np.pi,diffcrox1,label=r'$T_\chi=5$ MeV')
plt.plot(psi_ls2/np.pi,diffcrox2,label=r'$T_\chi=50$ MeV')
plt.plot(psi_ls3/np.pi,diffcrox3,label=r'$T_\chi=100$ MeV')
plt.yscale('log')
plt.legend()
plt.xlim(-0.02,0.25+0.02)
plt.ylim(3.3e-36)
plt.text(0.23,2e-35,r'$m_V\sim m_\chi=1$ MeV',horizontalalignment='center')
plt.text(0.23,8e-36,r'$g_V=g_D=10^{-3}$',horizontalalignment='center')

plt.xlabel(r'Scattering angle $\psi~[\pi]$')
plt.ylabel(r'$\frac{d\sigma_{\chi\nu}}{d\cos\psi}~~[{\rm cm^2\,rad}]$')
plt.savefig('../plots/diffcrox_xnu.pdf',bbox_inches='tight')
```
